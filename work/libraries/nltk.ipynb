Samples take from http://www.nltk.org/book/

```python
from nltk.book import *
```

```python
%matplotlib inline
```

```python
# get monty python text, included in nltk
month_python = text6
```

```python
# show frequencies based on offset in text
month_python.dispersion_plot(['coconut', 'shrubbery', 'knight','Arthur','Holy'])
```

```python
# build the distribution
f_dist = FreqDist(month_python)
```

```python
# show most common wods
f_dist.most_common(10)
```

```python
# show 10 most common alpha numerics, lowercased
al_nums = list((item.lower() for item in (month_python) if (item.isalnum())))
f_dist = FreqDist(al_nums)
f_dist.most_common(10)
```

```python
# total number of 'words'
len(month_python)
```

```python
# plot cumulative counts of most frequent words
from pylab import rcParams
rcParams['figure.figsize'] = 15, 5
f_dist.plot(50, cumulative=True)
```

```python
# show 'hapaxes'; words that only occur once
f_dist.hapaxes()
```

```python
# collocations show words that go together
month_python.collocations()
```
